```python
import tweepy as tw

API_KEY = 'jpsMh76nAhLYUraW4Wgb2zOqk'
API_SECRET = 'DmdhPi5XMKmVq3u4TChpxeRV1mUI1G5zRQ3Q1kuZIQOpVN8Qaw'
ACCESS_TOKEN = '1481137707166224384-FAsoE7Ez7wgvr67Hjc2dbezlzvJMuP'
ACCESS_TOKEN_SECRET = 'h4i2DfzjyebX4b8dnjzhpfSvDhDKcHieIjy619eFfaFWZ'

auth = tw.OAuthHandler(API_KEY, API_SECRET)
auth.set_access_token(ACCESS_TOKEN, ACCESS_TOKEN_SECRET)
api = tw.API(auth, wait_on_rate_limit=True)




from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import numpy as np
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic')
!pip install konlpy 
import re
from konlpy.tag import Komoran

```

```python
# 'ì˜í™”'ê°€ í¬í•¨ëœ íŠ¸ìœ—ì˜ í…ìŠ¤íŠ¸ 100ê°œ ì¶”ì¶œ
tweets = api.search(q='ì˜í™”', count=100)
documents = [tweet.text for tweet in tweets]   # 100ê°œ íŠ¸ìœ—ì˜ í…ìŠ¤íŠ¸
documents[0:2]


# ì¶”ì¶œëœ 100ê°œ ë¬¸ì„œ(íŠ¸ìœ—)ë“¤ì˜ ì „ì²˜ë¦¬

def preprocess(text):
    text = ' '.join(re.findall(r'[ê°€-í£]+', text))                    # í•œêµ­ì–´ ë‹¨ì–´ë§Œ ê°€ì ¸ì˜´ (ì´ ì‘ì—… ìˆ˜í–‰í•˜ì§€ ì•Šìœ¼ë©´ unicode ì—ëŸ¬ ë°œìƒ ê°€ëŠ¥)
    komoran = Komoran()
    tokens = komoran.nouns(text)                                      # ëª…ì‚¬ë§Œ ê°€ì ¸ì˜´
    tokens = [t for t in tokens if re.fullmatch(r'[ê°€-í£]{2,}', t)]   # 2ê¸€ì ì´ìƒ ë‹¨ì–´ ì„ íƒ
    return ' '.join(tokens)


documents_preprocessed = [preprocess(text) for text in documents]     # ì „ì²˜ë¦¬ëœ ë¬¸ì„œ
documents_preprocessed[0:2]


# ì¶”ì¶œëœ 100ê°œ ë¬¸ì„œë“¤ì˜ ë¬¸ì„œë‹¨ì–´í–‰ë ¬ ë§Œë“¤ê¸°
count_vectorizer = CountVectorizer(min_df=2, max_df=0.5)   # 2ê°œ ì´ìƒ ì „ì²´ ë¬¸ì„œì˜ 50% ì´í•˜ì˜ ë¬¸ì„œì—ì„œ ì¶œí˜„í•œ ë‹¨ì–´ë§Œ ê³ ë ¤
feat_vect = count_vectorizer.fit_transform(documents_preprocessed)
feature_names = count_vectorizer.get_feature_names_out()
print('ë‹¨ì–´ ëª©ë¡:', feature_names)
print('ë¬¸ì„œë‹¨ì–´í–‰ë ¬ ëª¨ì–‘ (ë¬¸ì„œì˜ ìˆ˜, ë‹¨ì–´ì˜ ìˆ˜):', feat_vect.shape)
print(feat_vect.toarray())


# í† í”½ëª¨ë¸ (LDA) ì ìš©í•˜ê¸°
lda = LatentDirichletAllocation(n_components=3)  # í† í”½ ìˆ˜ëŠ” 3ê°œë¡œ ì„¤ì •
lda.fit(feat_vect)
topic_word_dist = lda.components_ / lda.components_.sum(axis=1)[:, np.newaxis]    # í† í”½ë³„ ë‹¨ì–´ë¶„í¬
doc_topic_dist = lda.transform(feat_vect)                                         # ë¬¸ì„œë³„ í† í”½ë¶„í¬


# í† í”½ë³„ ì¤‘ìš”ë‹¨ì–´ ì¶œë ¥
n_top_words = 5    # ì¤‘ìš” ë‹¨ì–´ 5ê°œ
for i, topic_dist in enumerate(topic_word_dist):
    sorted_topic_dist = np.argsort(topic_dist)
    topic_words = np.array(feature_names)[sorted_topic_dist]
    topic_words = topic_words[:-n_top_words-1:-1]
    print("Topic", str(i+1), topic_words)

```

```python
# í† í”½ë³„ ë‹¨ì–´ë¶„í¬
i = 1    # í† í”½ ì¸ë±ìŠ¤ (ì¸ë±ìŠ¤ëŠ” 0ë¶€í„° ì‹œì‘í•¨ì— ìœ ì˜, í† í”½ 1: i=0)
topic_dist = topic_word_dist[i,:]
sorted_topic_dist = np.argsort(topic_dist)
topic_words = np.array(feature_names)[sorted_topic_dist]
topic_words = topic_words[:-n_top_words-1:-1]
print("Topic", str(i+1), topic_words)
plt.bar(topic_words, topic_word_dist[i,sorted_topic_dist][:-n_top_words-1:-1])
plt.xticks(rotation=90)
plt.title("Topic " + str(i+1))
plt.ylabel("Topic-specific Word Distribution")
plt.show()
```

```python
import tweepy as tw

API_KEY = 'jpsMh76nAhLYUraW4Wgb2zOqk'
API_SECRET = 'DmdhPi5XMKmVq3u4TChpxeRV1mUI1G5zRQ3Q1kuZIQOpVN8Qaw'
ACCESS_TOKEN = '1481137707166224384-FAsoE7Ez7wgvr67Hjc2dbezlzvJMuP'
ACCESS_TOKEN_SECRET = 'h4i2DfzjyebX4b8dnjzhpfSvDhDKcHieIjy619eFfaFWZ'

auth = tw.OAuthHandler(API_KEY, API_SECRET)
auth.set_access_token(ACCESS_TOKEN, ACCESS_TOKEN_SECRET)
api = tw.API(auth, wait_on_rate_limit=True)




from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import numpy as np
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic')
!pip install konlpy 
import re
import nltk
nltk.download('omw-1.4')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')
from nltk.corpus import stopwords
stopwords_en = stopwords.words('english')
```

```python
# 'movie'ê°€ í¬í•¨ëœ íŠ¸ìœ—ì˜ í…ìŠ¤íŠ¸ 100ê°œ ì¶”ì¶œ
tweets = api.search(q='movie', count=100)
documents = [tweet.text for tweet in tweets]   # 100ê°œ íŠ¸ìœ—ì˜ í…ìŠ¤íŠ¸
documents[0:2]


##ì¶”ì¶œëœ 100ê°œ ë¬¸ì„œ(íŠ¸ìœ—)ë“¤ì˜ ì „ì²˜ë¦¬
def preprocess(text):
    text = ' '.join(re.findall(r'[a-zA-Z_-]+', text)) 
    text = text.lower()
    tokens = nltk.word_tokenize(text)
    lemmatizer = nltk.stem.WordNetLemmatizer()
    irrelevant = ["rt", "co", "wa", "ak", "cr", "amp", 'http']
    tokens = [lemmatizer.lemmatize(t) for t in tokens]
    tokens = [t for t in tokens if t not in stopwords_en + irrelevant]
    tokens = [t for t in tokens if re.fullmatch('[a-z_-]+', t)]
    return ' '.join(tokens)

documents_preprocessed = [preprocess(text) for text in documents]     # ì „ì²˜ë¦¬ëœ ë¬¸ì„œ
documents_preprocessed[0:2]


## ì¶”ì¶œëœ 100ê°œì˜ ë¬¸ì„œë“¤ì˜ ë¬¸ì„œë‹¨ì–´í–‰ë ¬ ë§Œë“¤ê¸°
count_vectorizer = CountVectorizer(min_df=2, max_df=0.5)       # 2ê°œ ì´ìƒ ì „ì²´ ë¬¸ì„œì˜ 50% ì´í•˜ì˜ ë¬¸ì„œì—ì„œ ì¶œí˜„í•œ ë‹¨ì–´ë§Œ ê³ ë ¤
feat_vect = count_vectorizer.fit_transform(documents_preprocessed)
feature_names = count_vectorizer.get_feature_names_out()
print('ë‹¨ì–´ ëª©ë¡:', feature_names)
print('ë¬¸ì„œë‹¨ì–´í–‰ë ¬ ëª¨ì–‘ (ë¬¸ì„œì˜ ìˆ˜, ë‹¨ì–´ì˜ ìˆ˜):', feat_vect.shape)
print(feat_vect.toarray())


## í† í”½ëª¨ë¸(LDA) ì ìš©í•˜ê¸°
lda = LatentDirichletAllocation(n_components=3)  # í† í”½ ìˆ˜ëŠ” 3ê°œë¡œ ì„¤ì •
lda.fit(feat_vect)
topic_word_dist = lda.components_ / lda.components_.sum(axis=1)[:, np.newaxis]    # í† í”½ë³„ ë‹¨ì–´ë¶„í¬
doc_topic_dist = lda.transform(feat_vect)                                         # ë¬¸ì„œë³„ í† í”½ë¶„í¬


## í† í”½ë³„ ì¤‘ìš”ë‹¨ì–´ ì¶œë ¥
n_top_words = 5    # ì¤‘ìš” ë‹¨ì–´ 5ê°œ
for i, topic_dist in enumerate(topic_word_dist):
    sorted_topic_dist = np.argsort(topic_dist)
    topic_words = np.array(feature_names)[sorted_topic_dist]
    topic_words = topic_words[:-n_top_words-1:-1]
    print("Topic", str(i+1), topic_words)
```
