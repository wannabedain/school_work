# 네트워크

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()                                                       # 네트워크생성
G.add_nodes_from([1,2,3,4])                                          # 네트워크에 노드 추가
G.add_edges_from([(1,2), (1,3), (3,4), (2,3)])                       # 네트워크에 엣지 추가
nx.draw(G, with_labels=True, node_size=400, node_color='skyblue')   # 네트워크 그림 그리기
plt.show()
```

```python
# 네트워크 차수 구하기 (노드들의 차수 정보)

import numpy as np
degrees = dict(G.degree())
print(degrees)
print(degrees[4])
print(np.mean(list(degrees.values())))
```

```python
## 24p 연습문제

G = nx.Graph()                                                                       # Creating a network
G.add_nodes_from(['A', 'B', 'C', 'D', 'E'])                                          # Adding nodes
G.add_edges_from([('A','D'), ('B','C'), ('B','D'), ('C','D'), ('D','E')])            # Adding edges
nx.draw(G, with_labels=True, node_size=400, node_color='skyblue')                   # Drawing a Plot
plt.show()

degrees = dict(G.degree())
print('Degrees:', degrees)
```

```python
# 유향 네트워크 

DG = nx.DiGraph()   # Creating a directed network
DG.add_nodes_from(['A', 'B', 'C', 'D', 'E'])    # Adding nodes
DG.add_edges_from([('A','B'), ('B','A'), ('C','D'), ('A','D'), ('C','A'), ('E','A')])    # Adding edges
nx.draw(DG, with_labels=True, node_size=300, node_color='silver')    # Drawing a Plot
plt.show()

# 내차수와 외차수 
in_degrees = dict(DG.in_degree())
print(in_degrees)
out_degrees = dict(DG.out_degree())
print(out_degrees)
```

```python
# 랜덤 네트워크 

er = nx.erdos_renyi_graph(50, 0.05)
nx.draw(er, pos=nx.circular_layout(er), with_labels=True, node_color='skyblue')   # Drawing a Plot
plt.show()


# 차수 분포
degree_freq = nx.degree_histogram(er)
x = range(len(degree_freq))
plt.bar(x, degree_freq)           # Bar Plot
plt.show()
```

```python
# 척도 없는 네트워크 
ba = nx.barabasi_albert_graph(100, 1)  # 노드가 100개이고 평균 차수가 약 2인 네트워크
nx.draw(ba, with_labels=True, node_color='skyblue')   # Drawing a Plot
plt.show()


# 차수 분포    =>  차수가 매우 큰 노드(허브)들이 존재
degree_freq = nx.degree_histogram(ba)
x = range(len(degree_freq))
plt.bar(x, degree_freq)           # Bar Plot
plt.show()


#log-log 플롯    => 직선 형태를 보임
plt.loglog(x[1:], degree_freq[1:], 'go')      # log-log plot
plt.show()

```

# 노드 중심성 : 33p


```python
G = nx.Graph()                                                      # Creating a network
nodes = ['A','B','C','D','E','F','G','H','I','J','K','L','M','N','O']
edges = [('A','D'),('B','D'),('C','D'),('F','D'),('E','D'),('G','D'),('E','G'),('G','H'),('H','I'),
         ('I','L'),('J','L'),('K','L'),('M','L'),('N','L'),('O','L'),('I','J'),('N','O')]
G.add_nodes_from(nodes)
G.add_edges_from(edges)
pos = nx.spring_layout(G)  #
nx.draw(G, pos = pos, with_labels = True, node_size=400, node_color = 'gray')
```

## 차수 중심성

```python
nx.degree_centrality(G)    # 차수 / (전체 노드 수 - 1)     => D가 메우 넢음, L이 매우 높음
```

```python
print(nx.degree_centrality(G))    # degree / (number of nodes - 1)
nx.draw(G, pos=pos, with_labels=True, node_size=400, node_color=list(nx.degree_centrality(G).values()), cmap='coolwarm')   #coolwarm은 붉은 or 파랑으로 구분
plt.show()  
```

## 매개 중심성

```python
print(nx.betweenness_centrality(G))
nx.draw(G, pos=pos, with_labels=True, node_size=400, node_color=list(nx.betweenness_centrality(G).values()), cmap='coolwarm')   
plt.show()
```

```python
## 40p 연습문제
# 유향 네트워크 생성, 내차수, 외차수, 차수


DG = nx.DiGraph()   # Creating a directed network
DG.add_nodes_from(['A', 'B', 'C', 'D', 'E'])    # Adding nodes
DG.add_edges_from([('A','B'), ('A','C'), ('A','D'), ('B','D'), ('C','D'), ('C','E'), ('D','E'), ('E','A')])    # Adding edges
nx.draw(DG, with_labels=True, node_size=300, node_color='silver')    # Drawing a Plot
plt.show()
in_degrees = dict(DG.in_degree)
out_degrees = dict(DG.out_degree)
degrees = dict(DG.degree)
print('내차수',in_degrees)
print('외차수',out_degrees)  #활동적인 애가 누군지
print('차수',degrees)
```

```python
## 내차수 중심성, 외차수 중심성, 차수 중심성, 매개 중심성

print('내차수 중심성:', nx.in_degree_centrality(DG)) # 내차수 / (노드 수 -1)  => 5개 노드 이기 때문에! 나누기4
print('외차수 중심성:', nx.out_degree_centrality(DG)) # 내외차수 / (노드 수 -1)
print('차수중심성:', nx.degree_centrality(DG)) # 차수 / (노드 수 -1)
print('매개중심성:', nx.betweenness_centrality(DG))
```

# 커뮤니티 탐지

```python
from networkx.algorithms import community

def get_color_from_partition(partition, G):
    color = dict()
    for idx in range(len(partition)):
        for node in partition[idx]:
            color[node] = idx
    node_color = [color[node] for node in G.nodes()]
    return node_color
```

```python
# Greedy modularity maximization 방법
partition = list(community.greedy_modularity_communities(G))
print(partition)
nx.draw(G, pos=pos, with_labels=True, node_size=400, node_color=get_color_from_partition(partition, G), cmap='coolwarm')
plt.show()
```

```python
# Semi-synchronous label propagation 방법
partition = list(community.label_propagation_communities(G))
print(partition)
nx.draw(G, pos=pos, with_labels=True, node_size=400, node_color=get_color_from_partition(partition, G), cmap='coolwarm')
plt.show()
```

```python
## 46p 연습문제 : 노드들을 그룹화(커뮤니티 탐지)

# Fluid communities 방법
partition = list(community.asyn_fluidc(G, k=2))   # k: 커뮤니티 개수
print(partition)
nx.draw(G, pos=pos, with_labels=True, node_size=400, node_color=get_color_from_partition(partition, G), cmap='coolwarm')
plt.show()
```
